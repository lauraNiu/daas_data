```python
import random
def get_random_color():
    # 生成随机颜色
    color = "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
    return color
```

```python
import folium

# 提取轨迹数据
traj_data = traj_person1[traj_person1['date']==20231119]
# 该天的每次出行起终点都是紧密联结在一起的
print(traj_data)
```

```python
# 创建Folium地图对象，以第一个出行的起点为中心
m = folium.Map(location=[traj_data.iloc[0]['phi_o'], traj_data.iloc[0]['lambda_o']], zoom_start=20)

# 遍历轨迹数据，绘制每次出行的实线和虚线
for i in range(len(traj_data)):
    # 当前出行的起点和终点
    start_coords = [traj_data.iloc[i]['phi_o'], traj_data.iloc[i]['lambda_o']]
    end_coords = [traj_data.iloc[i]['phi_d'], traj_data.iloc[i]['lambda_d']]
    # 生成随机颜色
    random_color = get_random_color()
    # 使用随机颜色绘制PolyLine
    # 绘制起点到终点的实线
    folium.PolyLine([start_coords, end_coords], color=random_color, weight=2.5, opacity=1).add_to(m)

    # 在起点和终点放置标记
    folium.Marker(start_coords, icon=folium.Icon(color='green'),tooltip=traj_data.iloc[i]['t_start']).add_to(m)
    # folium.Marker(end_coords, icon=folium.Icon(color='red'),popup=traj_data.iloc[i]['seqid']).add_to(m)

    # 如果不是最后一次出行，绘制终点到下一次出行起点的虚线
    if i <len(traj_data) -1:
        print('i')
        next_start_coords = [traj_data.iloc[i+1]['phi_o'], traj_data.iloc[i+1]['lambda_o']]
        print(end_coords, next_start_coords)
        folium.PolyLine([end_coords, next_start_coords], color="blue", weight=2.5, opacity=1, dash_array='5, 5').add_to(m)
    else:
        print('结束')
# 显示地图
m
```
